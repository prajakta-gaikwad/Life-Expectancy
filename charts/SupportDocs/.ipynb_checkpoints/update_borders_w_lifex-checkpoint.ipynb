```python
import json
import os
import pandas as pd
# Load JSON
filepath = os.path.join("../data", "old", "countryBorders.geojson")
with open(filepath) as jsonfile:
    borders_geojson = json.load(jsonfile)
    
df_lifeex = pd.read_csv("../data/life-expectancy.csv")
df_lifex = df_lifeex.rename(index=str, columns={"Entity":"Country","Life expectancy (Clio-Infra up to 1949; UN Population Division for 1950 to 2015)": "Life-ex"})
```

```python

df_lifex.replace('United States','United States of America', inplace=True)
```

```python
for x in range(255):
    jsoncntry = borders_geojson['features'][x]['properties']['ADMIN']
    
    if len(df_lifex.loc[df_lifex['Country'] == jsoncntry])>0:
        df_lifex_filt = df_lifex.loc[df_lifex['Country'] == jsoncntry]
        df_lifex2 = df_lifex_filt[['Year','Life-ex']]
        

        ret_lifex2 = df_lifex2.to_json(orient='split')
        
        json_acceptable_string = ret_lifex2.replace("'", "\"")
        d = json.loads(json_acceptable_string)
        d.pop('columns', None)
        d.pop('index', None)
        
        borders_geojson['features'][x]['properties']['LifeX']={}
        borders_geojson['features'][x]['properties']['LifeX'] =d
        print(str(x) + " " + jsoncntry)
    else:
        print(str(x) + ' ---->' + jsoncntry)
        
        
```

```python
filepath2 = os.path.join("..", "data", "cb_updated.geojson")
with open(filepath2,"w") as jsonFile:
    json.dump(borders_geojson, jsonFile)
    
```
